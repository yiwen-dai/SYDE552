# Import and Process Data

```python
train_img_path = '/images/train/'
test_img_path = '/images/test/'

# Dataframes
train_df = pd.read_csv('fairface_label_train.csv')
test_df = pd.read_csv('fairface_label_test.csv')
```

```python
# Remove service_test column (unused: see issue https://github.com/joojs/fairface/issues/9) 
train_df = train_df.drop('service_test', axis=1)
test_df = test_df.drop('service_test', axis=1)


# Encode age, gender, and race
#    and keep map for each: {encoded_label: actual_label}
le = LabelEncoder()

age_encoded = le.fit_transform(train_df['age'])
train_df['age_encoded'] = age_encoded
age_map = {i: l for i, l in enumerate(le.classes_)}
age_encoded = le.fit_transform(test_df['age'])
test_df['age_encoded'] = age_encoded

gender_encoded = le.fit_transform(train_df['gender'])
train_df['gender_encoded'] = gender_encoded
gender_map = {i: l for i, l in enumerate(le.classes_)}
gender_encoded = le.fit_transform(test_df['gender'])
test_df['gender_encoded'] = gender_encoded

# Sample data from train_df
train_df.head()
```

```python
# Sample mapping from encoded age range to actual age range
print(age_map)
```

```python
# Custom dataset to get image and custom label
class ImageDataset(Dataset):
    def __init__(self, data_frame, root_dir, label_col, transform=None):
        self.data_frame = data_frame
        self.root_dir = root_dir
        self.label_col = label_col
        self.transform = transform 
    
    def __len__(self):
        # Return the length of the dataset
        return len(self.data_frame)
    
    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        # image
        img_path = os.path.join(self.root_dir, self.data_frame.iloc[idx]['file'])
        image = Image.open(img_path)
        image = (np.array(image))
        image = image.transpose(2,0,1)
        if self.transform:
            image = self.transform(image)
        
        # label
        category = self.data_frame.iloc[idx][self.label_col+'_encoded']
    
        return image, category
```

```python
# Set up diff dataframes for diff training and testing purposes
homogenous_train_df = train_df[train_df['race'] == 'White']
heterogenous_train_df = train_df
homogenous_test_df = test_df[test_df['race'] == 'White']
heterogenous_test_df = test_df
```

# Model

```python
# Get cpu or gpu device for training.
device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using {device} device")
```

```python
# Output the number of output neurons needed for final layer (depends on training data)
def get_n_outputs(label_col):
    match label_col:
        case 'age':
            return len(age_map)
        case 'gender':
            return len(gender_map)
```

```python
# Define custom CNN for image classification
class CNN(nn.Module):
    def __init__(self, n_output):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 32, 5)
        self.conv2 = nn.Conv2d(32, 64, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.fc1 = nn.Linear(179776, 512)    # input (3, 224, 224) -> (220, 220, 32)
        self.fc2 = nn.Linear(512, 128)     
        self.fc3 = nn.Linear(128, n_output)

        # define proportion of neurons to dropout to avoid overfitting
        self.pool_dropout = nn.Dropout(0.15)
        self.layer_dropout = nn.Dropout(0.15)

        # transforms
        self.transform = transforms.ToTensor()


    def forward(self, x):
        x = x.float()
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = self.pool_dropout(x)

        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = self.pool_dropout(x)

        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = self.layer_dropout(x)
        x = F.relu(self.fc2(x))
        x = self.layer_dropout(x)
        x = F.log_softmax(self.fc3(x), dim=1)
        return x
```

# Train and Test Functions

```python
# Train function
def train(dataloader, model, optimizer, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    train_loss, correct = 0, 0
    model.train()
    
    for batch, (images, categories) in enumerate(dataloader):
        images, categories = images.to(device), categories.to(device)
        
        optimizer.zero_grad()
        pred = model(images)
        loss = loss_fn(pred, categories)
        loss.backward()
        optimizer.step()
        _, pred = torch.max(pred, 1)
    
        train_loss += loss.item()
        correct += sum(x == y for x, y in zip(pred, categories))

    average_train_loss = train_loss / num_batches
    accuracy = correct / size
    return accuracy, average_train_loss
```

```python
# Test
def test(dataloader, model, loss_fn):
    print('1')
    size = len(dataloader.dataset)
    print('2')    
    num_batches = len(dataloader)
    
    model.eval()
    
    print('3')
    test_loss, correct = 0, 0
    
    with torch.no_grad():
        for batch, (images, categories) in enumerate(dataloader):
            print('4')
            images, categories = images.to(device), categories.to(device)

            print('5')
            pred = model(images)
            print('6')
            _, pred = torch.max(pred, 1)
            
            print('7')

            test_loss += loss_fn(torch.Tensor(pred).to(device), categories.float().to(device)).item()
            correct += sum(x == y for x, y in zip(pred, categories))
                
    average_test_loss = test_loss / num_batches
    accuracy = correct / size
    return accuracy, average_test_loss
```

```python
BATCH_SIZE = 64

def train_model(train_set, model, loss_fn, optim, epochs=3):
  for t in tqdm(range(epochs)):
      print(f"Epoch {t+1}:")
    
      # produce training set and validation set
      train_size = int(len(train_set) * 0.9)
      test_size = int(len(train_set) - train_size)
      split_train, split_valid = random_split(train_set, [train_size, test_size])
      train_loader = DataLoader(split_train, batch_size=BATCH_SIZE, shuffle=True)
      valid_loader = DataLoader(split_valid, batch_size=BATCH_SIZE, shuffle=True)
      
      # train
      acc, loss = train(train_loader, model, optim, loss_fn)
      print(f"\tAverage train accuracy: {100*acc:0.2f}%\t Avg train loss: {loss:>4f} \t")
      
    
      # test
      test_accuracy, test_loss = test(valid_loader, model, loss_fn)
      print(f"\tTest accuracy: {100*test_accuracy.item():0.2f}% \t Test loss: {test_loss.item():>4f}")

  return 
```

# Training

```python
# Create model for predicting age range of face based on homogenous data
homogenous_age_model = CNN(get_n_outputs('age')).to(device)
homogenous_age_optim = torch.optim.Adam(homogenous_age_model.parameters(), lr=0.001)
homogenous_age_loss = nn.CrossEntropyLoss()
```

```python
# Train and test data for age
homogenous_age_train_set = ImageDataset(data_frame=homogenous_train_df,
                                        root_dir= './images/',
                                        label_col='age')
homogenous_age_test_set = ImageDataset(data_frame=homogenous_test_df,
                                        root_dir='./images/',
                                        label_col='age')
```

```python
train_model(homogenous_age_train_set, homogenous_age_model, homogenous_age_loss, homogenous_age_optim)
```
